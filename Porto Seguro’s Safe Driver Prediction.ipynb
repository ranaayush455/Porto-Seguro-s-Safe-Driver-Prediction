```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
driver=pd.read_csv('train.csv')
driver
```

```python
driver.isnull().sum()
```

```python
driver.drop(columns='id',inplace=True)
```

```python
driver.dtypes
```

```python
y = driver['target']
driver.drop(columns='target',inplace=True)

from sklearn.model_selection import train_test_split 


X_train,X_test,y_train,y_test = train_test_split( driver,y,test_size=0.2,random_state=24)
```

```python

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
X_train_res,y_train_res = sm.fit_sample(X_train,y_train.ravel())

print('After OverSampling,the shape of train_X:{}'.format(X_train_res.shape))
print('After OverSampling ,the shape of train_y:{}\n'.format(y_train_res.shape))

print("After OverSampling,count of label '1':{}".format(sum(y_train_res==1)))
print("After OverSampling,count of label '0':{}".format(sum(y_train_res==0)))
```

```python
y_train_res
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=60,random_state=48)
rf.fit(X_train_res,y_train_res)
prediction=rf.predict(X_test)
prediction
```

```python
from sklearn.linear_model import LogisticRegression
LogR= LogisticRegression()
LogR.fit(X_train_res,y_train_res)
predicted=LogR.predict(X_test)
predicted
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,predicted)
```

```python
from sklearn.metrics import confusion_matrix, precision_score,accuracy_score
```

```python
confusion_matrix(y_test,predicted)
```

```python
precision_score(y_test,predicted)
```

```python
accuracy_score(y_test,predicted)
```
